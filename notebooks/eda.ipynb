# 2022 AIS Program Data

```python
ais_path = "../data/datadotgov_ais22_programs.csv"
ais_data = pd.read_csv(ais_path)
ais_data.head()
```

```python
# some exciting features here - like lat-lons for up to 10 operating locations
# features for beneficiaries (cols 5-33)
# only a third of these records have a link
ais_data.info()
```

```python
ais_data.columns
```

```python
# let's take a look at a subset of columns
cols_of_interest = [
    'ABN',
    'Charity Name',
    'Program name',
    'Operating Location 1',
    'Operating Location 1 lat/long',
    'Charity weblink'
]

# let's also filter for a specific beneficiary - homeless people
ais_data.loc[ais_data['People at risk of homelessness/ people experiencing homelessness']=='Y',cols_of_interest].sample(10)
```

observations:
- Lats and lons are pipe delimited, to 7 decimal places or precision, and some are 0.0000000 values
- many programs do not have a weblink - possible that this exists in the general AIS statement data (with abn potentially as join key)

ais_data.duplicated(subset='ABN',keep=False)
